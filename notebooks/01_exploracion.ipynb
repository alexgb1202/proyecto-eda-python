```python
# 1 Importamos las librerias que necesitamos
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None) # Para ver todas las columnas
```

```python
# 1.1 Importamos bank-additional (csv)
b_add = pd.read_csv('../Data/raw/bank-additional.csv', delimiter=',', index_col=0) # bank-additional.csv

# Importamos customer-details.xlsx (excel) con varias hojas
df_2012 = pd.read_excel('../Data/raw/customer-details.xlsx', sheet_name="2012")
df_2013 = pd.read_excel('../Data/raw/customer-details.xlsx', sheet_name="2013")
df_2014 = pd.read_excel('../Data/raw/customer-details.xlsx', sheet_name="2014")

# Convertimos las 3 hojas excel en un solo df
df_customer = pd.concat(objs=[df_2012, df_2013, df_2014], axis=0)
df_customer.drop(columns='Unnamed: 0', inplace=True) # Eliminamos la columna de indice incremental 'Unnamed: O'
```
